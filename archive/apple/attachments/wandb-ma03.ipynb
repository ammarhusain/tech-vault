```python
## WANDB Trial with OpenAI API
import openai
import datetime
import wandb, os
from wandb.sdk.data_types.trace_tree import Trace

os.environ["WANDB_API_KEY"] = "local-f92f2077a71582772ae4feee107979b308081d39"

openai.api_key = "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"

wandb.init(project="openai-trace-example")


# define your conifg
model_name = "gpt-3.5-turbo"
temperature = 0.7
system_message = "You are a helpful assistant that always replies in 3 concise bullet points using markdown."

queries_ls = [
  "What is the capital of France?",
  "How do I boil an egg?" * 10000,  # deliberately trigger an openai error
  "What to do if the aliens arrive?" 
]

for query in queries_ls:
    messages=[
      {"role": "system", "content": system_message},
      {"role": "user", "content": query}
    ]

    start_time_ms = datetime.datetime.now().timestamp() * 1000
    try:
      print(f"Calling openai {messages}")
      response = openai.ChatCompletion.create(model=model_name,
                                              messages=messages,
                                              temperature=temperature
                                              )   

      end_time_ms = round(datetime.datetime.now().timestamp() * 1000)  # logged in milliseconds
      status="success"
      status_message=None,
      response_text = response["choices"][0]["message"]["content"]
      token_usage = response["usage"].to_dict()
      
    
    except Exception as e:
      end_time_ms = round(datetime.datetime.now().timestamp() * 1000)  # logged in milliseconds
      status="error"
      status_message=str(e)
      response_text = ""
      token_usage = {}

    # create a span in wandb
    root_span = Trace(
          name="root_span",
          kind="llm",  # kind can be "llm", "chain", "agent" or "tool"
          status_code=status,
          status_message=status_message,
          metadata={"temperature": temperature,
                    "token_usage": token_usage, 
                    "model_name": model_name},
          start_time_ms=start_time_ms,
          end_time_ms=end_time_ms,
          inputs={"system_prompt": system_message, "query": query},
          outputs={"response": response_text},
          )
    
    # log the span to wandb
    root_span.log(name="openai_trace")
```

```python
## WANDB Trial with LlamaCpp
from llama_cpp import Llama

llm = Llama(model_path='/Users/ammarh/j595/llama.cpp/models/llama-2/llama-2-13b-chat.ggmlv3.q8_0.bin')

os.environ["WANDB_API_KEY"] = "local-f92f2077a71582772ae4feee107979b308081d39"

openai.api_key = "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"

wandb.init(project="llama-trace-example")


# define your conifg
model_name = "llama-2-13b-chat.ggmlv3.q8_0.bin"
temperature = 0.7
max_tokens = -1
system_message = "You are a helpful assistant that always replies in 3 concise bullet points using markdown."

queries_ls = [
  "What is the capital of France?",
  "How do I boil an egg?",
  "What to do if the aliens arrive?" 
]

for query in queries_ls:
    messages=[
      {"role": "system", "content": system_message},
      {"role": "user", "content": query}
    ]

    start_time_ms = datetime.datetime.now().timestamp() * 1000
    try:
      prompt = messages[0]["content"] + "\n" +  messages[1]["content"]
      print(f"Calling llama {prompt}")
      out = llm.create_completion(prompt=prompt, max_tokens=-1, temperature=temperature)   

      end_time_ms = round(datetime.datetime.now().timestamp() * 1000)  # logged in milliseconds
      status="success"
      status_message=None,
      response_text = out['choices'][0]['text']
      token_usage = "not set"
      
    
    except Exception as e:
      end_time_ms = round(datetime.datetime.now().timestamp() * 1000)  # logged in milliseconds
      status="error"
      status_message=str(e)
      response_text = ""
      token_usage = {}

    # create a span in wandb
    root_span = Trace(
          name="root_span",
          kind="llm",  # kind can be "llm", "chain", "agent" or "tool"
          status_code=status,
          status_message=status_message,
          metadata={"temperature": temperature,
                    "token_usage": token_usage, 
                    "model_name": model_name},
          start_time_ms=start_time_ms,
          end_time_ms=end_time_ms,
          inputs={"prompt": prompt},
          outputs={"response": response_text},
          )
    
    # log the span to wandb
    root_span.log(name="llama_trace")
```
