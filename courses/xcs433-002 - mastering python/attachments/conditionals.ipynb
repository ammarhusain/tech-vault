### if else statement

```python
if washing_machine_start:
    print('machine started')
else:
    print('machine stopped')
```

### if elif statement

```python
if washing_machine_start:
    print('machine started')
elif not door_closed:
    print('door is not closed')
else:
    print('machine stopped')
```

```python
start_pressed = True
door_closed = True
washing_machine_start = start_pressed and door_closed

if washing_machine_start:
    print('machine started')
elif not door_closed:
    print('door is not closed')
elif not start_pressed:
    print('start not pressed')
else:
    print('machine stopped')
```

### more conditional examples

```python
attempts = 0
last_score = 0

if attempts < 3 and last_score > 70:
    print('you passed the exam')
```

```python
attempts = 1
last_score = 80
if attempts < 3 and last_score > 70:
    print('you passed the exam')
```

```python
attempts = 4
last_score = 80
if attempts < 3 and last_score > 70:
    print('you passed the exam')
elif attempts >= 3:
    print('too many attempts')
```

```python
attempts = 2
last_score = 60
if attempts < 3 and last_score > 70:
    print('you passed the exam')
elif attempts >= 3:
    print('too many attempts')
else:
    print('score too low')
```

### multiple choice example

```python
# multiple choice
print('What is 5 + 10?')
print('a) 5')
print('b) 10')
print('c) 15')
print('d) 510')

selection = input("Enter Your Selection: ")
if selection == 'a':
    print('incorrect')
elif selection == 'b':
    print('incorrect')
elif selection == 'c':
    print('correct')
elif selection == 'd':
    print('incorrect')
else:
    print('select something between a and d')
```
