##1.
(20 points) Write code to write to a text file.
- write 20 random numbers, one number per line
- numbers should range from 0 to 1000
- use the random.randint function to generate the random integers
- for usage on randint, visit https://docs.python.org/3/library/random.html#functions-for-integers (Links to an external site.)

And finally, make sure your code handles some file exceptions, such as file not found, etc...

```


import random
try:
  f = open('drive/My Drive/1.txt', 'w')
  for i in range(20):
    f.write(str(random.randint(0,1000)) + "\n")
except OSError as oe:
  print(oe)

f.close()
```

##2.
(20 points) Write code to read from the text file you created from #1.
- Look for the line with the largest number
- Print BOTH the line number and the largest number itself
And finally, make sure your code handles file exceptions, such as file not found, etc...

```
try:
  f = open('drive/My Drive/1.txt', 'r') 
  txt = f.read()
except OSError as oe:
    print(oe)
f.close()

vals = txt.split()
print(f"Max val: {max(vals)} , line # {vals.index(max(vals))}")
```

(20 points)

Write code to write to a binary file.

- It should first write the length of a string of your choice
- Followed by the string itself
- Assume that the length is between 0 to 255 characters (one byte)
- Assume that the string encoding is utf-8.

Next, write code to read from this binary file.

- It should get the length of the string
- Then use this number to read the right amount of bytes from the file
- Finally, it should print the length of the string and the string itself

```
st = "Perilous delight in swimming in shallow waters."
with open('drive/My Drive/4.txt', 'wb') as f:
  f.write((len(st)).to_bytes(1, byteorder='little'))
  f.write(bytes(st, encoding='utf-8'))


with open('drive/My Drive/4.txt', 'rb') as f:
  # Read the first byte for length.
  lgt = 0
  bt = f.read(1)
  lgt = lgt.from_bytes(bt, byteorder='little')
  print("boom",lgt)
  txt = str(f.read(lgt), encoding='utf-8')
  print(f"String length: {lgt} ... string: '{txt}'")
```
