## 1. (10 points) 
Write a factorial function. A factorial is the product of an integer and all the integers below it.
For example, the factorial of 5 (5!) is 120 = (1 * 2 * 3 * 4 * 5)

````
def factorial(n):
  if n == 1:
    return 1
  return n * factorial(n-1)

print (factorial(5))
````

 Write your own version of the range() function. However, instead of returning a range type, it should return a list. Also, do not use the range function.
It should handle all of the parameters:

If passed one argument, it should return a list starting from 0 to the given argument minus one.

For example, myrange(5) should return:
```
[0, 1, 2, 3, 4]
```
If passed two arguments, it should return a list starting from the first argument to the second argument minus one.

For example, myrange(2, 6) should return:
```
[2, 3, 4, 5]
```
And finally, if passed three arguments, the third argument is the step count. It should do the same thing as the previous case but skip by the given number of steps.

For example, myrange(3, 30, 4) should return:
```
[3, 7, 11, 15, 19, 23, 27]
```
And finally, it should handle negative steps. If the third argument is negative, it should return a range of integers going downwards. (Note that the first argument is higher than the second argument.)

For example, myrange(10, 5, -1) should return:
```
[10, 9, 8, 7, 6]
```

````
def range_func(*args):
  s = 0
  stp = 1
  if (len(args) == 1):
    e = args[0]
  elif (len(args) ==  2):
    s = args[0]
    e = args[1]
  elif (len(args) ==  3):
    s = args[0]
    e = args[1]
    stp = args[2]
  else:
    return []

  out = []
  if stp > 0:
    while s < e:
      out.append(s)
      s += stp
  elif stp < 0:
    while s > e:
      out.append(s)
      s += stp
  else:
    raise ValueError("Step size must be non-zero.")

  return out

print(range_func(8, 4,-1))

print(range_func(7))

print(range_func(3, 9,0))
````

##3. (20 points)
Write a generator function that generates even numbers.
It should have one optional integer argument (n).
If the argument is not specified, it will generate an infinite number of even numbers.
If the argument is specified, then it should generate up to n even numbers.

Example:
```
# even_gen1 has an infinite number of iterations
even_gen1 = even_generator()
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
# even_gen2 should last only 4 iterations
even_gen2 = even_generator(4)
for i in even_gen2:
    print(f'even_gen2: {i}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
```
Example Output:
```
even_gen1: 0
even_gen1: 2
even_gen1: 4
even_gen2: 0
even_gen2: 2
even_gen2: 4
even_gen2: 6
even_gen1: 6
even_gen1: 8
even_gen1: 10
```

````
def even_generator(n=-1):
  num = 0
  while n != 0:
    yield num
    num += 2
    n -= 1

print(list(even_generator(9)))

# even_gen1 has an infinite number of iterations
even_gen1 = even_generator()
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
# even_gen2 should last only 4 iterations
even_gen2 = even_generator(4)
for i in even_gen2:
    print(f'even_gen2: {i}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
print(f'even_gen1: {next(even_gen1)}')
````

##4. (20 points) 
Write a decorator function that will clean up a string argument.
It should limit the string argument to 10 characters or less.
It should also convert the string into all lower case.

Example:
```
@cleanup_input
def print_input(s):
    return f'You entered the following input [{s}]'

@cleanup_input
def login_exists(s):
    logins = ['abc', 'root', 'user', 'admin', 'guest']
    if s in logins:
        return True
    else:
        return False

print(print_input('0123456789ABC')) # should limit string to 10 characters
print(print_input('Test')) # should print all lower case
print(login_exists('root')) # should be True
print(login_exists('Root')) # should be True
print(login_exists('gary')) # should be False
```

Output:
```
You entered the following input [0123456789]
You entered the following input [test]
True
True
False
 ```

````
def cleanup_input(f):
  def decorated(str):
    str = str[0:10].lower()
    return f(str)
  return decorated
  

@cleanup_input
def print_input(s):
    return f'You entered the following input [{s}]'
 
@cleanup_input
def login_exists(s):
    logins = ['abc', 'root', 'user', 'admin', 'guest']
    if s in logins:
        return True
    else:
        return False
 
print(print_input('0123456789ABC')) # should limit string to 10 characters
print(print_input('Test')) # should print all lower case
print(login_exists('root')) # should be True
print(login_exists('Root')) # should be True
print(login_exists('gary')) # should be False


````

##5. (10 points)
Using map(), write code to convert a list of numbers (both integers and floating point numbers) into a list of integers. If the number is less than 0, replace it with a 0.
```
[1.0, 3.14, 2.5, -7.39, 203.16, 9843, -1]
```
should be converted to:
```
[1, 3, 2, 0, 203, 9843, 0]
```

````
conv = lambda x : int(x) if x > 0 else 0

print(list(map(conv, [1.0, 3.14, 2.5, -7.39, 203.16, 9843, -1])))
````

##6. (10 points) 
Using filter() and a lambda function, write code to filter a list of strings and leave it as a list of strings that contain only digits.

For example:
```
['123', 'abc', '991', 'zero', '0', 'Fifteen', '024']
```
should result in:
```
['123', '991', '0', '024']
```

````
func = lambda x: x.isdigit()

print(list(filter(func, ['123', 'abc', '991', 'zero', '0', 'Fifteen', '024'])))
````
