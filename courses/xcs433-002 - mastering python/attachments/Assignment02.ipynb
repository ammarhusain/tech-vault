[1, 2, 3, 4, 5, 6, 7, 8]

To:

[2, 4, 6, 8, 10, 12]
 

```
lst = [1, 2, 3, 4, 5, 6, 7, 8]
lst.remove(1)
lst.remove(3)
lst.remove(5)
lst.remove(7)
lst.extend([10,12])
print(lst)
```

(15 points) Write a for loop that will multiply all the numbers of a list together. Assume all of the items in the list are numbers.

```
lst = [1, 2, 3, 4, 5, 6, 7, 8]
mul = 1
for l in lst:
  mul = mul*l

print(mul)
```

(10 points) Write a while loop that will multiply all the numbers of a list together (same as above but using a while loop)

```
lst = [1, 2, 3, 4, 5, 6, 7, 8]
mul = 1
i = 0
while i < len(lst):
  mul *= lst[i]
  i+=1
print(mul)
```
