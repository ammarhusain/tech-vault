## 2 
Add the following methods to the Circle class:

circumference() - returns the circumference of the circle (pi * diameter)

diameter() - returns the diameter of the circle


```
import math

class Shape:
  def __init__(self, x, y):
    self.x = x
    self.y = y
                     
class Circle(Shape):
  def __init__(self, radius):
    super().__init__(0, 0)
    self.radius = radius

  def area(self):
    return (self.radius ** 2) * math.pi
  def circumference(self):
    return math.pi * self.diameter()
  def diameter(self):
    return 2*self.radius


c = Circle(5)
print(c.area())
print(c.circumference())
print(c.diameter())
```

## 3

An ellipse is to a rectangle as a circle is to a square.  So an ellipse has two radiuses instead of one radius in a circle. It must include the area() method (PI * A * B) 



```
class Ellipse(Shape):
  def __init__(self, radius_a, radius_b):
    super().__init__(0, 0)
    self.radius_a = radius_a
    self.radius_b = radius_b

  def area(self):
    return self.radius_a * self.radius_b * math.pi

e = Ellipse(3,4)
print(e.area())
```

## 4
Add new properties to the Shape class and update the subclasses as required.

Add the following properties:

- an integer id

- a string name

Also, add the following methods:

Modify the __init__ method to automatically assign an id to the shape.

Write the following methods:

- get_id() - return the id

- set_name(string) - sets the name to the given string

- get_name() - returns the name

```
class Shape:
  def __init__(self, x, y, id):
    self.x = x
    self.y = y
    self.id = id
    self.name = ""
  def get_id(self):
    return self.id
  def set_name(self, name):
    self.name = name
  def get_name(self):
    return self.name

class Circle(Shape):
  def __init__(self, radius, id=1):
    super().__init__(0, 0, id)
    self.radius = radius
  def area(self):
    return (self.radius ** 2) * math.pi
  def circumference(self):
    return math.pi * self.diameter()
  def diameter(self):
    return 2*self.radius

class Ellipse(Shape):
  def __init__(self, radius_a, radius_b, id=1):
    super().__init__(0, 0, id)
    self.radius_a = radius_a
    self.radius_b = radius_b
  def area(self):
    return self.radius_a * self.radius_b * math.pi


c = Circle(5, id=7)
print(c.area())
print(c.circumference())
print(c.diameter())
print(c.get_id())
c.id = 8
print(c.get_id())
print(c.get_name())
c.set_name("circle")
print(c.get_name())

e = Ellipse(3,4, id=9)
print(e.area())
print(e.get_id())
e.id = 17
print(e.get_id())
print(e.get_name())
e.set_name("ellipse")
print(e.get_name())
```
