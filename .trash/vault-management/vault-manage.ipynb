```python
import re, os, hashlib, fnmatch

def get_files_to_index(rootdir, searchstr='publish: true'):
    files_to_index = []
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            if fnmatch.fnmatch(file, '*.md') == True:
                filepath = os.path.join(subdir, file)
                if fnmatch.fnmatch(filepath,'*/.trash/*') == False:
                    with open(filepath, 'r') as f:
                        if searchstr in f.read():
                            files_to_index.append(filepath)
    return files_to_index

def replace_line_in_file(file_path, orig_str="publish: \n", swap_str="publish: true\n"):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with open(file_path, 'w') as file:
        for line in lines:
            if orig_str in line:
                line = line.replace(orig_str, swap_str)
            file.write(line)

        
def remove_line_in_file(file_path, orig_str="publish: \n"):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with open(file_path, "w") as file:
        for line in lines:
            if orig_str not in line:
                file.write(line)
```

# Replace string in place   

```python
import os

def replace_string_in_files(directory, target, replacement):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".md"):  # Only process Markdown files
                file_path = os.path.join(root, file)
                
                # Read the file content
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                
                # Replace the target string
                if target in content:
                    updated_content = content.replace(target, replacement)
                    
                    # Write the updated content back to the file
                    with open(file_path, 'w', encoding='utf-8') as f:
                        f.write(updated_content)
                    print(f"Updated: {file_path}")

if __name__ == "__main__":
    # Set the path to your Obsidian vault's Daily folder
    daily_folder = "/Users/ammarh/Documents/second-brain-2/Daily"
    
    # Define the target and replacement strings
    target_string = "#### NOTES MODIFIED IN THE LAST WEEK"
    replacement_string = "# NOTES MODIFIED IN THE LAST WEEK"
    
    # Run the replacement function
    replace_string_in_files(daily_folder, target_string, replacement_string)
```
